Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are the visual shapes that represent the data on a plot. Examples of glyphs include circles, lines, squares, triangles, and more. You can add glyphs to a Bokeh plot using the various glyph methods available in Bokeh.

```python
p = figure(plot_width=400, plot_height=400, title="My Plot")

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

p.circle(x, y, size=10, color="navy", alpha=0.5)
output_file("plot.html")
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
p = figure(plot_width=400, plot_height=400, title="My Plot")

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
p.circle(x, y, size=10, color="navy", alpha=0.5, legend_label = 'Data')

p.xaxis.axis_label = "X-axis Label"
p.yaxis.axis_label = "Y-axis Label"
p.xaxis.ticker = [1, 2, 3, 4, 5]
p.yaxis.ticker = [2, 4, 6, 8]
p.xaxis.major_label_orientation = "vertical"

p.legend.title = "Legend Title"
p.legend.location = "bottom_right"
p.legend.label_text_font_size = "14pt"

output_file("plot.html")
show(p)
```
