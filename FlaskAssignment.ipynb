## Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight and flexible web application framework written in Python. It is designed to make it easy to build web applications quickly and efficiently.

Advantages of Flask framework:

Lightweight and easy to use: Flask is a lightweight framework and easy to use. It does not require any complicated setup or installation, and it has a simple syntax that makes it easy to learn and use.

Flexibility: Flask provides developers with the flexibility to choose the components and tools they want to use. This allows developers to customize their web applications and build them exactly the way they want.

Modular Design: Flask is designed in a modular way, which means that developers can use only the features they need. This helps keep the codebase small and efficient.

Built-in Development Server: Flask comes with a built-in development server that makes it easy to test and debug web applications.

Large Community: Flask has a large community of developers who contribute to its development and offer support. This means that developers can get help and advice from other developers when they need it.

## Q3. What is App routing in Flask? Why do we use app routes?

App routing in Flask is a mechanism that maps a URL to a Python function that handles the request for that URL. In other words, app routing in Flask is used to determine which Python function should be executed when a user makes a request to a specific URL.

To create an app route in Flask, you use the @app.route decorator. This decorator is used to associate a URL endpoint with a Python function that will handle the request for that endpoint.


We use app routes in flask for several reason :

Handling user requests: App routing allows us to handle user requests to specific URLs and execute the necessary Python code to respond to those requests.

Organizing code: App routing allows us to organize our code into separate functions, each of which handles a specific URL. This makes our code easier to read and maintain.

Building RESTful APIs: App routing is an important component of building RESTful APIs in Flask. Each app route can correspond to a specific RESTful endpoint, allowing us to create a robust API for our application.

## 4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
## Company Name: ABC Corporation
## Location: India
## Contact Detail: 999-999-9999

```python
from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return 'Welcome to ABC Corporation'

@app.route('/')
def home():
    company_name = 'ABC Corporation'
    location = 'India'
    contact_detail = '999-999-9999'
    
    return f'Company Name: {company_name}<br>Location: {location}<br>Contact Detail: {contact_detail}'

if __name__ == '__main__':
    app.run(host="0.0.0.0")
```

## Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

In Flask, the url_for() function is used for URL building. It generates a URL to the specified endpoint with the given arguments.

```python
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def index():
    return 'Hello, World!'

@app.route('/user/<username>')
def profile(username):
    return f'Hello, {username}!'

with app.test_request_context():
    # Generate a URL for the index() function
    index_url = url_for('index')
    print(index_url)  # Output: /

    # Generate a URL for the profile() function with the username argument set to 'John'
    profile_url = url_for('profile', username='John')
    print(profile_url)  # Output: /user/John
```
